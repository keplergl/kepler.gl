# Empty & Null-Geometry GeoDataFrames

Demonstrates that `KeplerGl` can accept GeoDataFrames that are empty or
contain only null geometries.  This is useful when an analysis pipeline
produces no spatial results — the layer still appears on the map,
signalling that the analysis ran but found nothing.

```python
import pandas as pd
import geopandas
```

## Empty GeoDataFrame (zero rows)

```python
empty_gdf = geopandas.GeoDataFrame(
    {"name": pd.Series([], dtype="str")},
    geometry=geopandas.GeoSeries([], crs="EPSG:4326"),
)
display(empty_gdf)
```

## All-null geometries (rows exist, but every geometry is None)

```python
null_geom_gdf = geopandas.GeoDataFrame(
    {"label": ["a", "b", "c"]},
    geometry=[None, None, None],
)
display(null_geom_gdf)
```

## Load into kepler.gl

```python
import keplergl
w = keplergl.KeplerGl(height=500)
w
```

```python
w.add_data(data=empty_gdf, name="empty")
```

```python
w.add_data(data=null_geom_gdf, name="null_geom")
```
