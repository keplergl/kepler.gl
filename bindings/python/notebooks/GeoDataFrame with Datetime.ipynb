# GeoDataFrame with Datetime Columns

Demonstrates that `KeplerGl` handles GeoDataFrames containing
`datetime` (and other non-JSON-native) columns when adding data
and exporting to HTML.

By default, `save_to_html` uses `default=str` when serializing
GeoDataFrame data to GeoJSON, so datetime values are automatically
converted to strings.  You can override this with the `json_encoder`
parameter.

```python
import pandas as pd
import geopandas
```

## Create a GeoDataFrame with datetime columns

```python
df = pd.DataFrame({
    "City": ["Buenos Aires", "Brasilia", "Santiago", "Bogota", "Caracas"],
    "Country": ["Argentina", "Brazil", "Chile", "Colombia", "Venezuela"],
    "Latitude": [-34.58, -15.78, -33.45, 4.60, 10.48],
    "Longitude": [-58.66, -47.91, -70.66, -74.08, -66.86],
    "Timestamp": pd.to_datetime([
        "2024-01-15 08:00",
        "2024-02-20 09:30",
        "2024-03-10 11:00",
        "2024-04-05 14:15",
        "2024-05-22 16:45",
    ]),
})

gdf = geopandas.GeoDataFrame(
    df, geometry=geopandas.points_from_xy(df.Longitude, df.Latitude)
)
display(gdf)
```

## Load into kepler.gl

```python
import keplergl

w = keplergl.KeplerGl(height=500)
w
```

```python
w.add_data(data=gdf, name="cities")
```

## Export to HTML

The default `json_encoder=str` converts datetime values automatically.

```python
w.save_to_html(file_name="datetime_map.html", read_only=True)
```

## Custom json_encoder

You can pass a custom function to control how non-serializable types
are encoded.  For example, use `isoformat()` for ISO 8601 strings:

```python
w.save_to_html(
    file_name="datetime_map_iso.html",
    read_only=True,
    json_encoder=lambda o: o.isoformat() if hasattr(o, "isoformat") else str(o),
)
```
