## Apply a config with polygon color scale

You can set layer styles, basemap, and more through the `config` dict.
The `config` structure mirrors what `w1.config` returns — you can interactively
style your map in the widget, then copy `w1.config` into your code.

Below we load the GeoJSON zip code polygons and apply a config that:
- Uses a **geojson** layer with a quantize color scale on `OBJECTID`
- Sets the basemap to **positron** (a light free basemap, no Mapbox token needed)

```python
polygon_config = {
    'version': 'v1',
    'config': {
        'visState': {
            'layers': [{
                'type': 'geojson',
                'config': {
                    'dataId': 'zip_codes',
                    'label': 'SF Zip Codes',
                    'isVisible': True,
                    'visConfig': {
                        'opacity': 0.8,
                        'filled': True,
                        'stroked': True,
                        'thickness': 1,
                        'colorRange': {
                            'name': 'Sunrise 8',
                            'type': 'sequential',
                            'category': 'Uber',
                            'colors': [
                                '#194266', '#355C7D', '#63617F', '#916681',
                                '#C06C84', '#D28389', '#E59A8F', '#F8B195'
                            ]
                        }
                    },
                    'columns': {'geojson': '_geojson'}
                },
                'visualChannels': {
                    'colorField': {'name': 'OBJECTID', 'type': 'integer'},
                    'colorScale': 'quantize'
                }
            }]
        },
        'mapStyle': {
            'styleType': 'positron'
        }
    }
}

w2 = keplergl.KeplerGl(
    height=500,
    data={'zip_codes': json_data},
    config=polygon_config,
    theme='light',
    app_name='SF Zip Codes'
)
w2
```

```python
# Export the styled polygon map
w2.save_to_html(file_name='sf_zip_styled.html')
```
