# Theme and App Name

This notebook demonstrates how to customize the kepler.gl UI theme and application name.

**`theme`** controls the kepler.gl UI appearance:
- `"light"` – light UI, pairs well with Mapbox light basemap styles
- `"dark"` – dark UI (default)
- `"base"` – base theme

**`app_name`** controls the title shown in the side panel header and the HTML `<title>` when exporting.

```python
import pandas as pd
import keplergl
```

```python
df = pd.DataFrame({
    'City': ['Buenos Aires', 'Brasilia', 'Santiago', 'Bogota', 'Caracas'],
    'Country': ['Argentina', 'Brazil', 'Chile', 'Colombia', 'Venezuela'],
    'Latitude': [-34.58, -15.78, -33.45, 4.60, 10.48],
    'Longitude': [-58.66, -47.91, -70.66, -74.08, -66.86]
})
```

## Default (dark theme)

```python
w_default = keplergl.KeplerGl(height=400, data={'cities': df})
w_default
```

## Light theme with custom app name

```python
w_light = keplergl.KeplerGl(
    height=400,
    data={'cities': df},
    theme='light',
    app_name='City Explorer'
)
w_light
```

## Export to HTML

Both `theme` and `app_name` carry over to `save_to_html`. You can also override them at export time.

```python
# Uses the widget's theme="light" and app_name="City Explorer"
w_light.save_to_html(file_name='light_theme_map.html')
```

```python
# Override theme and app_name at export time
w_default.save_to_html(
    file_name='custom_export.html',
    theme='light',
    app_name='My Custom Map'
)
```
